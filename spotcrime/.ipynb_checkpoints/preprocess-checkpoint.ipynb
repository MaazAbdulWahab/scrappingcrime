```python
import pandas as pd
```

```python
data=pd.read_csv('dataspotcrimeunprepared.csv')
```

```python
data.head()
```

```python
def removefirsttoken(dat):
    dat=dat.split(':')[1]
    return dat
```

```python
data['Type']=data['Field1'].apply(removefirsttoken)
data['Description']=data['Field3'].apply(removefirsttoken)
data['Address']=data['Field4'].apply(removefirsttoken)
```

```python
data.head()
```

```python
data['Field5_Link'][0]
```

```python
def parsedate(dat):
    dat=dat.split(':')
    dat=dat[1:]
    datreturn=dat[0]+':'+dat[1]
    return datreturn
```

```python
data['DateTime']=data['Field2'].apply(parsedate)
```

```python
data.head()
```

```python
def parselatitude(dat):
    dat=dat.split('=')[2]
    dat=dat.split(',')
    dat=dat[0]
    return dat
```

```python
def parselongitude(dat):
    dat=dat.split(' ')[1]
    
    return dat
```

```python
data['Latitude']=data['Field5_Link'].apply(parselatitude)
data['Longitude']=data['Field5_Link'].apply(parselongitude)
```

```python
data.head()
```

```python
data=data[['Type','Description','Address','DateTime','Latitude','Longitude']]
```

```python
data.to_csv('SpotCrimeNewYork.csv', index=False)
```
